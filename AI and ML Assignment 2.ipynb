```julia
# First add the package (only needed once)
using Pkg
Pkg.add("MLJModelInterface")

# Then use the packages (on separate lines)
using MLJModelInterface
using DataFrames
using StatsBase

# Drop SMILES column
df_clean = select(df, Not(:SMILES))

# Remove rows with any NaN or Inf values
df_clean = filter(row -> all(x -> !ismissing(x) && !isnan(x) && isfinite(x), row), eachrow(df_clean))

# Convert to matrix
X = Matrix(df_clean)

# Standardize (zscore each column manually if dims is not supported)
X_std = (X .- mean(X, dims=1)) ./ std(X, dims=1)
```

```julia
#K-Means Clustering
```

```julia
using Clustering

k = 3  # you can try different values
kmeans_result = kmeans(X_std', k)

# Cluster assignments
kmeans_labels = kmeans_result.assignments

# Optional: visualize with PCA
Pkg.add("MultivariateStats")
using MultivariateStats, Plots

X_pca = fit(PCA, X_std; maxoutdim=2) |> (p -> transform(p, X_std))
scatter(X_pca[:, 1], X_pca[:, 2], group=kmeans_labels, legend=false, title="K-Means Clustering")
```

```julia
#Agglomerative Clustering
```

```julia
using Clustering

# Use hierarchical clustering with Ward linkage
hier = hclust(pairwise(Euclidean(), X_std, dims=2), linkage=:ward)

# Choose the same number of clusters as k-means
agg_labels = cutree(hier, k)

# Visualize
scatter(X_pca[:, 1], X_pca[:, 2], group=agg_labels, legend=false, title="Agglomerative Clustering")
```

```julia
#Gaussian Mixture Model
```

```julia
using GaussianMixtures

gmm = GMM(k, X_std; method=:kmeans)
gmm_labels = assign(gmm, X_std)

# Visualize
scatter(X_pca[:, 1], X_pca[:, 2], group=gmm_labels, legend=false, title="Gaussian Mixture Clustering")
```

```julia
#Evaluating Clustering Quality
```

```julia
using MLJBase

silhouette_kmeans = silhouette_score(X_std, kmeans_labels)
silhouette_agg = silhouette_score(X_std, agg_labels)
silhouette_gmm = silhouette_score(X_std, gmm_labels)

println("Silhouette Score - KMeans: ", silhouette_kmeans)
println("Silhouette Score - Agglomerative: ", silhouette_agg)
println("Silhouette Score - GMM: ", silhouette_gmm)
```
