```julia
#Installing packages
using Pkg
Pkg.add("ARFFFiles")  
Pkg.add("DataFrames")
Pkg.add("CSV")
Pkg.add("DecisionTree")
Pkg.add("MLJ")
Pkg.add("MLJBase")
Pkg.add("Plots")
Pkg.add("StatsPlots")
Pkg.add("Statistics")
Pkg.add("Random")

# Loading the packages
using ARFFFiles 
using DataFrames, CSV
using DecisionTree
using MLJ, MLJBase
using Plots, StatsPlots
using Statistics
using Random
Random.seed!(123)  

# Loading the ARFF file 
data = ARFFFiles.load("C:/Users/ENekundi/Desktop/Memory 2025/MSC Data Science/Trends in artificial Intelligence and Machine Learning/Trends in AI and Machine Learning Group Assignment 2 - Final/ag_soybean.arff") 
df = DataFrame(data)

# Display basic information about the dataset
println("Dataset dimensions: ", size(df))
println("First few rows:")
first(df, 5)
```

```julia
#Data Cleaning
#Checking for missing values
println("Missing values per column:")
show(describe(df), allcols=true)

# Handling missing values
df = dropmissing(df)

# Convert categorical variables to appropriate types
# Assuming the class/target variable is the last column
target_col = names(df)[end]
X = df[:, 1:end-1]
y = df[:, end]

# Convert target to categorical
y = categorical(y)

# Split data into train and test sets (70% train, 30% test)
train, test = partition(eachindex(y), 0.7, shuffle=true, stratify=y)

# One-hot encode categorical features (if needed)
# This depends on your dataset - some ensemble methods can handle categorical features directly
```

```julia
# Ensemble 1: Bagging
# Defining the bagging model

using Pkg
Pkg.add("MLJ")
Pkg.add("MLJEnsembles")  # This package contains BaggingClassifier
Pkg.add("MLJBase")

# Loading the necessary packages
using MLJ, MLJBase, MLJEnsembles

# Defining the bagging model
bagging_model = BaggingClassifier(
    n_iterations=50,  # number of base models
    rng=123
)

# Make sure X and y are properly defined
# X should be your feature matrix/table
# y should be your target vector
# For example:
# X = select(df, Not(:target_column))
# y = df.target_column

# Wrap in machine
bagging_machine = machine(bagging_model, X, y)

# Training
fit!(bagging_machine, rows=train)

# Predictions
y_pred_bagging = predict_mode(bagging_machine, rows=test)

# Evaluating
accuracy_bagging = accuracy(y[test], y_pred_bagging)
precision_bagging = precision(y[test], y_pred_bagging)
recall_bagging = recall(y[test], y_pred_bagging)

println("Bagging Results:")
println("Accuracy: ", accuracy_bagging)
println("Precision: ", precision_bagging)
println("Recall: ", recall_bagging)
```

```julia
# Defining boosting model (using AdaBoost)
boosting_model = AdaBoostStumpClassifier(
    n_iterations=50,  # number of boosting iterations
    algorithm=:SAMME,  # boosting algorithm
    rng=123
)

# Wrap in machine
boosting_machine = machine(boosting_model, X, y)

# Train
fit!(boosting_machine, rows=train)

# Predictions
y_pred_boosting = predict_mode(boosting_machine, rows=test)

# Evaluate
accuracy_boosting = accuracy(y[test], y_pred_boosting)
precision_boosting = precision(y[test], y_pred_boosting)
recall_boosting = recall(y[test], y_pred_boosting)

println("Boosting Results:")
println("Accuracy: ", accuracy_boosting)
println("Precision: ", precision_boosting)
println("Recall: ", recall_boosting)
```

```julia
# Defining base models for stacking
models = [
    (@load DecisionTreeClassifier pkg=DecisionTree verbosity=0)(),
    (@load RandomForestClassifier pkg=DecisionTree verbosity=0)(),
    (@load LogisticClassifier pkg=MLJLinearModels verbosity=0)()
]

# Defining meta-model
meta_model = (@load LogisticClassifier pkg=MLJLinearModels verbosity=0)()

# Define stacking model
stacking_model = Stack(
    atomic_models=models,
    meta_model=meta_model,
    resampling=CV(nfolds=5)
)

# Wrap in machine
stacking_machine = machine(stacking_model, X, y)

# Train
fit!(stacking_machine, rows=train)

# Predictions
y_pred_stacking = predict_mode(stacking_machine, rows=test)

# Evaluate
accuracy_stacking = accuracy(y[test], y_pred_stacking)
precision_stacking = precision(y[test], y_pred_stacking)
recall_stacking = recall(y[test], y_pred_stacking)

println("Stacking Results:")
println("Accuracy: ", accuracy_stacking)
println("Precision: ", precision_stacking)
println("Recall: ", recall_stacking)
```

```julia
# Defining decision tree model
tree_model = (@load DecisionTreeClassifier pkg=DecisionTree verbosity=0)()

# Wrap in machine
tree_machine = machine(tree_model, X, y)

# Train
fit!(tree_machine, rows=train)

# Predictions
y_pred_tree = predict_mode(tree_machine, rows=test)

# Evaluate
accuracy_tree = accuracy(y[test], y_pred_tree)
precision_tree = precision(y[test], y_pred_tree)
recall_tree = recall(y[test], y_pred_tree)

println("Decision Tree Results:")
println("Accuracy: ", accuracy_tree)
println("Precision: ", precision_tree)
println("Recall: ", recall_tree)
```

```julia
# Compiling all metrics
metrics = DataFrame(
    Model = ["Bagging", "Boosting", "Stacking", "Decision Tree"],
    Accuracy = [accuracy_bagging, accuracy_boosting, accuracy_stacking, accuracy_tree],
    Precision = [precision_bagging, precision_boosting, precision_stacking, precision_tree],
    Recall = [recall_bagging, recall_boosting, recall_stacking, recall_tree]
)

println("Performance Comparison:")
show(metrics, allcols=true)
```

```julia
# For multiclass ROC, we need to binarize the problem for each class
# Here's an example for one-vs-rest approach for the first class
# (Note: This needs to be adapted based on your specific classes)

# Function to plot ROC for a model
function plot_roc(machine, y_true, test_rows, class_level)
    # Get predicted probabilities
    y_probs = predict(machine, rows=test_rows)
    
    # Binarize for the current class
    y_true_bin = y_true .== class_level
    y_scores = pdf.(y_probs, class_level)
    
    # Calculate ROC curve
    fpr, tpr, thresholds = roc(y_true_bin, y_scores)
    
    # Calculate AUC
    auc = trapz(fpr, tpr)
    
    # Plot
    plot(fpr, tpr, label="AUC = $(round(auc, digits=3))", legend=:bottomright)
    plot!([0, 1], [0, 1], linestyle=:dash, label="Random")
    title!("ROC Curve for $(class_level)")
    xlabel!("False Positive Rate")
    ylabel!("True Positive Rate")
end

# Example for one class (adjust based on your dataset)
class_to_plot = levels(y)[1] # first class

p1 = plot_roc(bagging_machine, y, test, class_to_plot)
p2 = plot_roc(boosting_machine, y, test, class_to_plot)
p3 = plot_roc(stacking_machine, y, test, class_to_plot)
p4 = plot_roc(tree_machine, y, test, class_to_plot)

plot(p1, p2, p3, p4, layout=(2,2), size=(800,600))
```
